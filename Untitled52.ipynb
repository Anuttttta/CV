```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞
!pip install kagglehub -q

import kagglehub
import os
import shutil
from pathlib import Path
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# 1. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞
print("–ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞")
path = kagglehub.dataset_download("phucthaiv02/butterfly-image-classification")
print("–ü—É—Ç—å –∫ –¥–∞—Ç–∞—Å–µ—Ç—É:", path)

dataset_root = Path(path)

# 2. –°–æ–±–∏—Ä–∞–µ–º –í–°–ï –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏–∑ –∫–æ—Ä–Ω–µ–≤—ã—Ö –ø–∞–ø–æ–∫-–∫–ª–∞—Å—Å–æ–≤
all_images = []
all_labels = []

# –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º –ø–æ–¥–ø–∞–ø–∫–∞–º –≤ –∫–æ—Ä–Ω–µ ‚Äî –∫–∞–∂–¥–∞—è = –æ—Ç–¥–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å
for class_dir in dataset_root.iterdir():
    if class_dir.is_dir():
        class_name = class_dir.name
        # –ò—â–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤–Ω—É—Ç—Ä–∏ –ø–∞–ø–∫–∏ –∫–ª–∞—Å—Å–∞
        for img_path in class_dir.glob("*"):
            if img_path.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']:
                all_images.append(str(img_path))
                all_labels.append(class_name)

print(f" –í—Å–µ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –Ω–∞–π–¥–µ–Ω–æ: {len(all_images)}")


if len(all_images) == 0:
    raise FileNotFoundError(" –ù–µ –Ω–∞–π–¥–µ–Ω–æ –Ω–∏ –æ–¥–Ω–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è")

# 3. –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É 70/15/15
new_dataset_root = Path("/content/butterfly_dataset_70_15_15")
shutil.rmtree(new_dataset_root, ignore_errors=True)
(new_dataset_root / "train").mkdir(parents=True)
(new_dataset_root / "val").mkdir(parents=True)
(new_dataset_root / "test").mkdir(parents=True)

# 4. –†–∞–∑–±–∏–≤–∞–µ–º —Å–Ω–∞—á–∞–ª–∞ –Ω–∞ train (70%) –∏ –æ—Å—Ç–∞–ª—å–Ω–æ–µ (30%)
X_train, X_rest, y_train, y_rest = train_test_split(
    all_images, all_labels,
    test_size=0.3,
    random_state=42,
    stratify=all_labels
)

# –ó–∞—Ç–µ–º –¥–µ–ª–∏–º "–æ—Å—Ç–∞–ª—å–Ω–æ–µ" –ø–æ–ø–æ–ª–∞–º ‚Üí 15% val, 15% test
X_val, X_test, y_val, y_test = train_test_split(
    X_rest, y_rest,
    test_size=0.5,
    random_state=42,
    stratify=y_rest
)

print(f"\n–†–∞–∑–º–µ—Ä—ã:")
print(f"  train: {len(X_train)}")
print(f"  val:   {len(X_val)}")
print(f"  test:  {len(X_test)}")

# 5. –§—É–Ω–∫—Ü–∏—è –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è
def copy_images(images, labels, split_name):
    split_dir = new_dataset_root / split_name
    for img_path, label in tqdm(zip(images, labels), total=len(images), desc=f"–ö–æ–ø–∏—Ä–æ–≤–∞–Ω–∏–µ ‚Üí {split_name}"):
        class_dir = split_dir / label
        class_dir.mkdir(exist_ok=True)
        shutil.copy2(img_path, class_dir / Path(img_path).name)

# 6. –ö–æ–ø–∏—Ä—É–µ–º
copy_images(X_train, y_train, "train")
copy_images(X_val, y_val, "val")
copy_images(X_test, y_test, "test")

# 7. –°–æ–∑–¥–∞—ë–º YAML –¥–ª—è YOLO
classes = sorted(set(all_labels))
yaml_content = f"""path: /content/butterfly_dataset_70_15_15
train: train
val: val
test: test
names: {classes}
"""

with open("/content/butterfly.yaml", "w") as f:
    f.write(yaml_content)

print("\n –î–∞—Ç–∞—Å–µ—Ç —É—Å–ø–µ—à–Ω–æ –ø–µ—Ä–µ—Å–æ–±—Ä–∞–Ω –≤ —Ñ–æ—Ä–º–∞—Ç–µ 70/15/15!")
print(" –ü—É—Ç—å:", new_dataset_root)
print(" YAML:", "/content/butterfly.yaml")
```

```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π
!pip install -q timm torch torchvision

# –ò–º–ø–æ—Ä—Ç—ã
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import timm
from pathlib import Path
from tqdm import tqdm

# –ü—É—Ç—å –∫ —É–∂–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–º—É –¥–∞—Ç–∞—Å–µ—Ç—É
new_dataset_root = Path("/content/butterfly_dataset_70_15_15")
train_dir = new_dataset_root / "train"
val_dir = new_dataset_root / "val"

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π
assert train_dir.exists(), "–ü–∞–ø–∫–∞ train –Ω–µ –Ω–∞–π–¥–µ–Ω–∞!"
assert val_dir.exists(), "–ü–∞–ø–∫–∞ val –Ω–µ –Ω–∞–π–¥–µ–Ω–∞!"

# –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏
img_size = 224
train_transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
train_dataset = datasets.ImageFolder(train_dir, transform=train_transform)
val_dataset = datasets.ImageFolder(val_dir, transform=val_transform)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤
assert train_dataset.classes == val_dataset.classes, "–ö–ª–∞—Å—Å—ã –≤ train –∏ val –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç!"

classes = train_dataset.classes
num_classes = len(classes)
print(f"–ù–∞–π–¥–µ–Ω–æ –∫–ª–∞—Å—Å–æ–≤: {num_classes}")

# DataLoader
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

# –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {device}")

# –ú–æ–¥–µ–ª—å EfficientNet-B0
model = timm.create_model('efficientnet_b0', pretrained=True, num_classes=num_classes)
model.to(device)

# –û–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä –∏ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5)

# –§—É–Ω–∫—Ü–∏—è –æ–±—É—á–µ–Ω–∏—è –æ–¥–Ω–æ–π —ç–ø–æ—Ö–∏
def train_one_epoch():
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for images, labels in tqdm(train_loader, desc="–û–±—É—á–µ–Ω–∏–µ"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    acc = correct / total
    return running_loss / len(train_loader), acc

# –§—É–Ω–∫—Ü–∏—è –≤–∞–ª–∏–¥–∞—Ü–∏–∏
def validate():
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc="–í–∞–ª–∏–¥–∞—Ü–∏—è"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    acc = correct / total
    return acc

# –û–±—É—á–µ–Ω–∏–µ
num_epochs = 30
best_acc = 0.0

for epoch in range(num_epochs):
    print(f"\n–≠–ø–æ—Ö–∞ [{epoch+1}/{num_epochs}]")
    train_loss, train_acc = train_one_epoch()
    val_acc = validate()

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}")

    if val_acc > best_acc:
        best_acc = val_acc
        torch.save(model.state_dict(), "/content/best_efficientnet_butterfly.pth")
        print(f"–ù–æ–≤–∞—è –ª—É—á—à–∞—è –º–æ–¥–µ–ª—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞! Val Acc: {best_acc:.4f}")

    scheduler.step(val_acc)

print(f"\n–û–±—É—á–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ. –õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å –Ω–∞ val: {best_acc:.4f}")
```

```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ (–µ—Å–ª–∏ –µ—â—ë –Ω–µ —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ)
!pip install -q timm torch torchvision matplotlib

# –ò–º–ø–æ—Ä—Ç—ã
import torch
import torch.nn as nn
from torchvision import datasets, transforms
import timm
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import random
import os

# === 1. –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ –∏ –¥–∞–Ω–Ω—ã—Ö ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
new_dataset_root = Path("/content/butterfly_dataset_70_15_15")
val_dir = new_dataset_root / "val"

# –ü—Ä–æ–≤–µ—Ä–∫–∞
assert val_dir.exists(), "–ü–∞–ø–∫–∞ val –Ω–µ –Ω–∞–π–¥–µ–Ω–∞!"

# –ó–∞–≥—Ä—É–∑–∫–∞ –∫–ª–∞—Å—Å–æ–≤ (–∏–∑ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –ø–∞–ø–æ–∫)
val_dataset = datasets.ImageFolder(val_dir)
classes = val_dataset.classes
num_classes = len(classes)

# –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏—è –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–∏
img_size = 224
val_transform = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏
model = timm.create_model('efficientnet_b0', pretrained=False, num_classes=num_classes)
model_path = "/content/best_efficientnet_butterfly.pth"
if not os.path.exists(model_path):
    raise FileNotFoundError("–ú–æ–¥–µ–ª—å –Ω–µ –Ω–∞–π–¥–µ–Ω–∞. –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –æ–±—É—á–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–∏–ª–æ—Å—å –∏ —Ñ–∞–π–ª —Å–æ—Ö—Ä–∞–Ω—ë–Ω.")

model.load_state_dict(torch.load(model_path, map_location=device))
model.to(device)
model.eval()

# === 2. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –ø—Ä–∏–º–µ—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π ===
# –°–æ–±–∏—Ä–∞–µ–º –¥–æ 9 —Å–ª—É—á–∞–π–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ val
sample_images = []
sample_labels = []

for _ in range(9):
    class_name = random.choice(classes)
    class_path = val_dir / class_name
    if class_path.exists():
        img_files = [f for f in class_path.iterdir() if f.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']]
        if img_files:
            img_path = random.choice(img_files)
            image = Image.open(img_path).convert("RGB")
            sample_images.append(image)
            sample_labels.append(class_name)

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
predictions = []
confidences = []

with torch.no_grad():
    for img in sample_images:
        input_tensor = val_transform(img).unsqueeze(0).to(device)
        output = model(input_tensor)
        probs = torch.softmax(output, dim=1)
        pred_idx = torch.argmax(probs, dim=1).item()
        confidence = probs[0][pred_idx].item()
        predictions.append(classes[pred_idx])
        confidences.append(confidence)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Å–µ—Ç–∫–∏ 3x3
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes = axes.ravel()

for i in range(len(sample_images)):
    axes[i].imshow(sample_images[i])
    color = "green" if predictions[i] == sample_labels[i] else "red"
    axes[i].set_title(
        f"–ò—Å—Ç–∏–Ω–Ω—ã–π: {sample_labels[i]}\n"
        f"–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–π: {predictions[i]}\n"
        f"–£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å: {confidences[i]:.2f}",
        color=color,
        fontsize=10
    )
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
